```python
import pandas as pd

# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```

## USCensus1990-2000

From the USCensus1990-2000 dataset, we use ward linkage as we don't know much about the theory behind the dataset. The dendogram shows most of the data have high similarity has the heights are mostly small. Step by step, the heirachical clustering group the data points together to form a cluster and keeps forming subcluster based on which point, cluster, sub-cluster have the most similarity (based on Euclidean distance) between them.
